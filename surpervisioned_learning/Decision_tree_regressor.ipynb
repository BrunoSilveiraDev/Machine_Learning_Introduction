```python
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np
from sklearn import tree
from sklearn.model_selection import train_test_split

uri = 'datasets/movies_multilinear_reg.csv'
dataset = pd.read_csv(uri)
dataset.head()
```

```python
x = dataset[dataset.columns[2:17]]
y = dataset[dataset.columns[17:]]
x_train,x_test, y_train, y_test  =  train_test_split(x,y)
```

```python
x_train = np.array(x_train).reshape(len(x_train), 15)
x_test = np.array(x_test).reshape(len(x_test), 15)
```

```python
model = tree.DecisionTreeRegressor(max_depth=5)
model.fit(x_train,y_train)

print('Score: %.2f%%' % (model.score(x_train, y_train)*100))
```
